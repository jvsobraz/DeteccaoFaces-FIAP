```
def one_hot(categorias, valor):
  categorias = list(set(categorias))
  indice = categorias.index(valor)
  vetor = np.zeros(len(categorias))
  vetor[indice] = 1

  return vetor
```

```
def normaliza_min(valores):
  # media = np.mean(valores)
  minimo = np.min(valores)
  maximo = np.max(valores)
  return (valores - minimo)/(maximo-minimo)
```

```
df.idade = normaliza_min(df.idade)
df.head()
```

```
df.dependentes = normaliza_min(df.dependentes)
df.head()
```

```
df.tarifa = normaliza_min(df.tarifa)
df.head()
```

```
df.parentes = normaliza_min(df.parentes)
df.head()
```
